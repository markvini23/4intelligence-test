### 1.3 Modelagem

```python
#ARIMA
model = auto_arima(time_series_df['valor'], trace=True, error_action='ignore', suppress_warnings=True)
model.fit(time_series_df['valor'])
forecast = model.predict(n_periods=12)
forecast = pd.DataFrame(forecast, columns=['valor'])

forecast['data'] = pd.date_range(start='8/1/2023', end='8/1/2024', freq='M')
forecast = forecast.set_index('data')

plt.plot(time_series_df['valor'], color='red')
plt.plot(forecast, color='blue')
```

```python
def optimize_SARIMA(parameters_list, d, D, s, exog):
    """
        Return dataframe with parameters, corresponding AIC of SARIMA model
    """
    
    results = []
    
    for param in tqdm_notebook(parameters_list):
        try: 
            model = SARIMAX(exog, order=(param[0], d, param[1]), seasonal_order=(param[2], D, param[3], s)).fit(disp=-1)
        except:
            continue
            
        aic = model.aic
        results.append([param, aic])
        
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q)x(P,Q)', 'AIC']
    #Sort in ascending order, lower AIC is better
    result_df = result_df.sort_values(by='AIC', ascending=True).reset_index(drop=True)
    
    return result_df
```

```python
p = range(0, 4, 1)
d = 1
q = range(0, 4, 1)
P = range(0, 4, 1)
D = 1
Q = range(0, 4, 1)
s = 1
parameters = product(p, q, P, Q)
parameters_list = list(parameters)


result_df = optimize_SARIMA(parameters_list, 1, 1, 4, time_series_df['valor'])
result_df
```

```python
result_df.sort_values(by='AIC')
```

```python
best_model = SARIMAX(time_series_df['valor'], order=(2, 1, 1), seasonal_order=(2, 1, 3, 3)).fit(dis=-1)
print(best_model.summary())
```

```python
time_series_df['sarima_model'] = best_model.fittedvalues
time_series_df['sarima_model'][:4+1] = np.NaN

forecast = best_model.predict(start=time_series_df.shape[0], end=time_series_df.shape[0] + 12)
forecast = pd.concat([time_series_df['sarima_model'], forecast])

plt.figure(figsize=(15, 7.5))
plt.plot(forecast, color='r', label='model')
plt.axvspan(time_series_df.index[-1], forecast.index[-1], alpha=0.5, color='lightgrey')
plt.plot(time_series_df['valor'], label='actual')
plt.legend()
plt.show()
```

### 2.3 Modelagem:

```python
def evaluate_model(y_valid, predict):
    #print scores
    accuracy = accuracy_score(y_valid, predict)
    recall = recall_score(y_valid, predict)
    f1 = f1_score(y_valid, predict) 
    print(f"Accuracy on validation set: {accuracy:.3f}")
    print(f"recall on validation set: {recall:.3f}")
    print(f"f1 score on validation set: {f1:.3f}")

    # #draw confusion matrix
    cm = confusion_matrix(y_valid, predict, labels=[0, 1])

    #plot confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
    disp.plot()
    plt.show()
```

```python
#Train, test, validation split
X = credit_df.drop(columns=['default.payment.next.month'])
y = credit_df['default.payment.next.month']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=23)

X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=23)
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)

#Logistic Regression
lr = LogisticRegression(random_state=23).fit(X_scaled, y_train)


X_valid_scaled = scaler.transform(X_valid)
predict = lr.predict(X_valid_scaled)

#Plot confusion and print scores
evaluate_model(y_valid, predict)
```

```python
scaler = preprocessing.StandardScaler().fit(X_train)
X_scaled = scaler.transform(X_train)

#Naive Bayes
gnb = GaussianNB().fit(X_scaled, y_train)


X_valid_scaled = scaler.transform(X_valid)
predict = gnb.predict(X_valid_scaled)

#Plot confusion and print scores
evaluate_model(y_valid, predict)
```

```python
#Catboost ~ Xgboost
cb = CatBoostClassifier()

cb.fit(X_train, y_train, verbose=0)
predict = cb.predict(X_valid)

#Plot confusion and print scores
evaluate_model(y_valid, predict)
```

```python
#Train with balanced dataset to catch the difference

smote = SMOTE()
x_smote, y_smote = smote.fit_resample(X,y)
X_train_smt, X_test_smt, y_train_smt, y_test_smt = train_test_split(x_smote, y_smote, test_size=0.15, random_state=23)

X_train_smt, X_valid_smt, y_train_smt, y_valid_smt = train_test_split(X_train_smt, y_train_smt, test_size=0.15, random_state=23)

rf = RandomForestClassifier()
xg = xgb.XGBClassifier()
cb = CatBoostClassifier()

cb.fit(X_train_smt, y_train_smt, verbose=0)
rf.fit(X_train_smt, y_train_smt)
xg.fit(X_train_smt, y_train_smt)

cb_pred = cb.predict(X_valid_smt)
xg_pred = rf.predict(X_valid_smt)
cb_pred = xg.predict(X_valid_smt)
```

```python
#Plot confusion and print scores
evaluate_model(y_valid_smt, cb_pred)
```

```python
#Plot confusion and print scores
evaluate_model(y_valid_smt, xg_pred)
```

```python
#Plot confusion and print scores
evaluate_model(y_valid_smt, cb_pred)
```

```python
#Best model
predict_test = xg.predict(X_test)
evaluate_model(y_test, predict_test)
```

### 2.3 Conclusão:

Com base nos testes realizados com diferentes modelos, foi possível observar que os modelos baseados em metodos de "Ensemble" alcançaram um melhor desempenho ao comparar métricas de Acurácia, revocação e F1 em conjunto com a matriz de confusão de resultado da predição na base de validação. Ao aplicar a predição do modelo Xgboost (modelo com melhor desempenho) foi possível obter um valor de F1 de 63% na base de teste.

Vale destacar alguns pontos a respeito da base de dados utilizada no problema: 
- A base possuia classes desbalanceadas, ou seja, o número de uma das classes era extreamente maior que a outra (proporção de ~1 para 3 ocorrências de uma classe para a outra), logo foi necessário aplicar métodos de sobreamostragem na base de dados para obter os melhores resultados, nesse caso, utilizei o método SMOTE para sobreamostragem sintetica na base de dados do problema;


- Para avaliar o desempenho dos modelos, a métrica considerada foi a F1, pois, essa métrica reflete a média ponderada entre acurácia e revocação. Em bases de dados desbalanceados a visualização da matriz de confusão em conjunto com a otimização da métrica F1 se mostra eficaz em conseguir uma performance adequada em problemas de classificação;


- Para um teste de desempenho adequado do modelo realizei a subdivisão da base de dados em treino, teste e validação. Essa técnica permite isolar os dados de teste na parte de escolha do melhor modelo a ser utilizado, evitando a contaminação do teste e consequentemente a ocorrência do efeito de sobreajuste do modelo.
